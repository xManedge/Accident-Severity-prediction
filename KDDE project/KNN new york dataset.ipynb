```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
import numpy as np 
```

```python
# Metric printing 
def metrics(ytrue,ypred):
    print(confusion_matrix(ytrue,ypred))
    print("the accuracy is ",accuracy_score(ytrue,ypred))
    #print("the precision is ",precision_score(ytrue,ypred))
    #print("the recall is ",recall_score(ytrue,ypred))
    print(classification_report(ytrue, ypred))
    return accuracy_score(ytrue,ypred)#, precision_score(ytrue,ypred), recall_score(ytrue,ypred)
```

```python
link = './accident dataset/new york accidents preprocessed.csv'
data = pd.read_csv(link)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    data.drop('Severity',axis=1),
    data.Severity,
    test_size=0.2
)
```

```python
param_grid = {
    'n_neighbors':[4,5,7,9,10], 
    'weights':['uniform', 'distance']
}
best = GridSearchCV(
    KNeighborsClassifier(n_jobs= -1), 
    param_grid=param_grid, 
    n_jobs=-1
)
```

getting best features according to Random forest

```python
xtrain = xtrain[['Zipcode', 'Distance(mi)', 'Street', 'Start_Lat','Start_Lng']]
xtest  = xtest[['Zipcode', 'Distance(mi)', 'Street', 'Start_Lat','Start_Lng']]
```

```python
best.fit(xtrain, ytrain)
```

```python
best.best_params_
```

```python
n_neighbors, weights = best.best_params_.values()
```

```python
weights
```

```python
knn = KNeighborsClassifier(
    n_neighbors= 9,#n_neighbors,
    #weights= weights,
    n_jobs = -1
)
```

```python
knn.fit(xtrain, ytrain)
```

```python
ypred = knn.predict(xtest)
```

```python
metrics(ytest, ypred)
```

# KNN on data with temp clusters

```python
link = './accident dataset/new york accident minimeans clustered.csv'
data = pd.read_csv(link)
data = data.drop('TimeTaken', axis=1)
data.head()
```

```python
data = data[['Zipcode', 'Distance(mi)', 'Street', 'Start_Lat','Start_Lng','tempClusters','Severity']]
data.head()
```

```python
#data = data.drop('TimeTaken',axis=1)
xtrain, xtest, ytrain, ytest = train_test_split(
    data.drop('Severity',axis=1),
    data.Severity,
    test_size=0.2
)
```

```python
knn = KNeighborsClassifier(
    n_neighbors=9, #n_neighbors
    #weights=weights
    n_jobs=-1,
)
```

```python
knn.fit(xtrain, ytrain)
```

```python
ypred = knn.predict(xtest)
```

```python
metrics(ytest, ypred)
```
